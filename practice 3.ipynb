```python
import math

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    else:
        return x / y

def power(x, y):
    return x ** y

def square_root(x):
    if x < 0:
        return "Error! Cannot compute square root of a negative number."
    else:
        return math.sqrt(x)

def factorial(x):
    if x < 0:
        return "Error! Cannot compute factorial of a negative number."
    elif x == 0:
        return 1
    else:
        return x * factorial(x - 1)

def main():
    print("Welcome to the Python Scientific Calculator!")
    print("Operations:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Power")
    print("6. Square Root")
    print("7. Factorial")
    print("0. Exit")

    while True:
        choice = input("Enter operation number: ")

        if choice == '0':
            print("Exiting calculator. Goodbye!")
            break

        if choice in ('1', '2', '3', '4', '5'):
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))

            if choice == '1':
                print("Result:", add(num1, num2))
            elif choice == '2':
                print("Result:", subtract(num1, num2))
            elif choice == '3':
                print("Result:", multiply(num1, num2))
            elif choice == '4':
                print("Result:", divide(num1, num2))
            elif choice == '5':
                print("Result:", power(num1, num2))

        elif choice == '6':
            num = float(input("Enter number: "))
            print("Result:", square_root(num))

        elif choice == '7':
            num = int(input("Enter number: "))
            print("Result:", factorial(num))

        else:
            print("Invalid input. Please enter a valid operation number.")

if __name__ == "__main__":
    main()
```

```python
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def make_sound(self):
        pass  # This will be overridden by subclasses


class Mammal(Animal):
    def __init__(self, name, species, fur_color):
        super().__init__(name, species)
        self.fur_color = fur_color

    def give_birth(self):
        return "Live birth"


class Reptile(Animal):
    def __init__(self, name, species, scale_type):
        super().__init__(name, species)
        self.scale_type = scale_type

    def lay_eggs(self):
        return "Lay eggs"


class Dog(Mammal):
    def __init__(self, name, breed, fur_color):
        super().__init__(name, "Canis lupus familiaris", fur_color)
        self.breed = breed

    def make_sound(self):
        return "Bark"


class Snake(Reptile):
    def __init__(self, name, species, scale_type):
        super().__init__(name, species, scale_type)

    def make_sound(self):
        return "Hiss"


def main():
    dog = Dog("Buddy", "Labrador Retriever", "Golden")
    snake = Snake("Slither", "Pythonidae", "Smooth")

    print(f"{dog.name} the {dog.breed} says: {dog.make_sound()}.")
    print(f"{snake.name} the {snake.species} says: {snake.make_sound()}.")


if __name__ == "__main__":
    main()
```

```python
class Animal:
    def __init__(self, name, species, habitat):
        self.name = name
        self.species = species
        self.habitat = habitat

    def move(self):
        return "Move in various ways"


class Mammal(Animal):
    def __init__(self, name, species, habitat, fur_color):
        super().__init__(name, species, habitat)
        self.fur_color = fur_color

    def give_birth(self):
        return "Give live birth"


class Reptile(Animal):
    def __init__(self, name, species, habitat, scale_type):
        super().__init__(name, species, habitat)
        self.scale_type = scale_type

    def lay_eggs(self):
        return "Lay eggs"


class Bird(Animal):
    def __init__(self, name, species, habitat, wingspan):
        super().__init__(name, species, habitat)
        self.wingspan = wingspan

    def fly(self):
        return "Fly"


class Fish(Animal):
    def __init__(self, name, species, habitat, water_type):
        super().__init__(name, species, habitat)
        self.water_type = water_type

    def swim(self):
        return "Swim"


class Dog(Mammal):
    def __init__(self, name, breed, fur_color):
        super().__init__(name, "Canis lupus familiaris", "Land", fur_color)
        self.breed = breed

    def make_sound(self):
        return "Bark"


class Snake(Reptile):
    def __init__(self, name, species, scale_type):
        super().__init__(name, species, "Land", scale_type)

    def make_sound(self):
        return "Hiss"


class Eagle(Bird):
    def __init__(self, name, species, wingspan):
        super().__init__(name, species, "Sky", wingspan)

    def make_sound(self):
        return "Screech"


class Goldfish(Fish):
    def __init__(self, name, species, water_type):
        super().__init__(name, species, "Freshwater", water_type)

    def make_sound(self):
        return "Glub glub"


def main():
    dog = Dog("Buddy", "Labrador Retriever", "Golden")
    snake = Snake("Slither", "Pythonidae", "Smooth")
    eagle = Eagle("Freedom", "Aquila chrysaetos", 200)
    goldfish = Goldfish("Bubbles", "Carassius auratus", "Coldwater")

    print(f"{dog.name} the {dog.breed} ({dog.species}) lives on {dog.habitat} and says: {dog.make_sound()}.")
    print(f"{snake.name} the {snake.species} lives on {snake.habitat} and says: {snake.make_sound()}.")
    print(f"{eagle.name} the {eagle.species} with a wingspan of {eagle.wingspan} cm lives in the {eagle.habitat} and says: {eagle.make_sound()}.")
    print(f"{goldfish.name} the {goldfish.species} lives in {goldfish.habitat} and says: {goldfish.make_sound()}.")


if __name__ == "__main__":
    main()
```

```python
# Function to classify animals into categories
def classify_animal(animal_name):
    mammals = ["dog", "cat", "elephant", "tiger", "human"]
    birds = ["sparrow", "eagle", "parrot", "penguin"]
    reptiles = ["snake", "lizard", "crocodile", "turtle"]
    amphibians = ["frog", "salamander", "newt"]
    fish = ["shark", "salmon", "goldfish"]

    if animal_name in mammals:
        return "mammal"
    elif animal_name in birds:
        return "bird"
    elif animal_name in reptiles:
        return "reptile"
    elif animal_name in amphibians:
        return "amphibian"
    elif animal_name in fish:
        return "fish"
    else:
        return "unknown"

# Function to display characteristics of each category
def display_characteristics(category):
    characteristics = {
        "mammal": "Mammals are warm-blooded vertebrates that have hair or fur and most give live birth. Examples: dog, cat, human.",
        "bird": "Birds are warm-blooded vertebrates with feathers, wings, and beaks. Most can fly. Examples: sparrow, eagle, parrot.",
        "reptile": "Reptiles are cold-blooded vertebrates that have scales and lay eggs. Examples: snake, lizard, crocodile.",
        "amphibian": "Amphibians are cold-blooded vertebrates that live both in water and on land. Examples: frog, salamander.",
        "fish": "Fish are cold-blooded vertebrates that live in water and have gills, fins, and scales. Examples: shark, salmon.",
        "unknown": "Unknown category. This animal is not in our database."
    }
    return characteristics.get(category, "Unknown category")

# Main function
def main():
    animals = ["dog", "sparrow", "snake", "frog", "shark", "dragon"]
    for animal in animals:
        category = classify_animal(animal)
        print(f"The {animal} is a {category}.")
        print(display_characteristics(category))
        print()  # Print a blank line for better readability

if __name__ == "__main__":
    main()
```
